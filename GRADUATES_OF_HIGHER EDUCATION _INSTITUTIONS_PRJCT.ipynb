```python
#Uzbekistan,Tashkent, City

graduates= pd.read_csv('my_project_graduates.csv')

fig = plt.figure(figsize=(11,4),facecolor='ivory')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)

fig.suptitle('Graduates in Uzbekistan during the decade'+" (2010-2019)", fontsize=16, fontweight='bold', color="darkblue")

plt.plot(graduates.Year, graduates.Uzbekistan,'.--', color="red",
        label='Uzbekistan')
plt.plot(graduates.Year, graduates.TashkentCity, '.--', color="lime",
        label='Tashkent City')

plt.xticks(graduates.Year)
plt.ylabel('Number of students',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")
plt.xlabel('Y e a r s',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")           
plt.legend(loc='upper right',fontsize=10, edgecolor="green", facecolor="white")
plt.grid(b=True, linestyle='dashdot')

plt.show()

#Bukhara, Jizzakh, Kashkadarya

graduates= pd.read_csv('my_project_graduates.csv')
plt.subplots(figsize=(11,4),facecolor='ivory')


plt.plot(graduates.Year, graduates.Bukhara,'.-', color="orangered",
        label='Bukhara')
plt.plot(graduates.Year, graduates.Jizzakh ,'.-', color="indigo",
        label='Jizzakh')
plt.plot(graduates.Year, graduates.Kashkadarya,'.-', color="magenta",
        label='Kashkadarya')

plt.xticks(graduates.Year)
plt.ylabel('Number of students',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")
plt.xlabel('Y e a r s',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")           
plt.legend(loc='upper right',fontsize=10, edgecolor="green", facecolor="white")
plt.grid(b=True, linestyle='dashdot')
plt.show()


#Navoi Namangan, Samarkand, Karakalpakistan

graduates= pd.read_csv('my_project_graduates.csv')
plt.subplots(figsize=(11,4),facecolor='ivory')

plt.plot(graduates.Year, graduates.Navoi,'.-', color="orange",
        label='Navoi')
plt.plot(graduates.Year, graduates.Namangan,'.-', color="turquoise",
        label='Namangan')
plt.plot(graduates.Year, graduates.Samarkand,'.-', color="darkblue",
        label='Samarkand')
plt.plot(graduates.Year, graduates.Karakalpakistan,'y.-',
        label='Karakalpakistan')
 
plt.xticks(graduates.Year)
plt.ylabel('Number of students',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")
plt.xlabel('Y e a r s',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")           
plt.legend(loc='upper center',fontsize=10, edgecolor="green", facecolor="white")
plt.grid(b=True, linestyle='dashdot')

plt.show()


#Surkhandarya, Syrdarya, Tashkent

graduates= pd.read_csv('my_project_graduates.csv')
plt.subplots(figsize=(11,4), facecolor='ivory')

plt.plot(graduates.Year, graduates.Surkhandarya,'.-', color='salmon',
        label='Surkhandarya')
plt.plot(graduates.Year, graduates.Syrdarya,'.-', color='sienna',
        label='Syrdarya')
plt.plot(graduates.Year, graduates.Tashkent,'.-', color='teal',
        label='Tashkent')

plt.xticks(graduates.Year)
plt.ylabel('Number of students',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")
plt.xlabel('Y e a r s',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")           
plt.legend(loc='upper right',fontsize=10, edgecolor="green", facecolor="white")
plt.grid(b=True, linestyle='dashdot')

plt.show()


#Andijan,Fergana,Khorezm

graduates= pd.read_csv('my_project_graduates.csv')
plt.subplots(figsize=(11,4),facecolor='ivory')

plt.plot(graduates.Year, graduates.Andijan,'.-', color='navy',
        label='Andijan')
plt.plot(graduates.Year, graduates.Fergana,'.-', color='orchid',
        label='Fergana')
plt.plot(graduates.Year, graduates.Khorezm,'.-', color='green',
        label='Khorezm')


plt.xticks(graduates.Year)
plt.ylabel('Number of students',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")
plt.xlabel('Y e a r s',fontdict={'fontweight':'bold', 'fontsize':13}, color="indigo")           
plt.legend(loc='upper right',fontsize=10, edgecolor="green", facecolor="white")
plt.grid(b=True, linestyle='dashdot')

plt.show()
```
